*How many random samples from no-subscription records do we need?*

```python
no_of_subs = train.y.value_counts()[1]
print('There are {} subscription records in the train data.'.format(no_of_subs))
```

*Randomly selecting **4203** no-subscription records*

```python
no_subs = train[train['y'] == 0]
subs = train[train['y'] == 1]
selected = no_subs.sample(no_of_subs)
selected.head()
```

*Concatenating both into a subsample data set with equal **y** distribution*

```python
selected.reset_index(drop=True, inplace=True)
subs.reset_index(drop=True, inplace=True)

subsample = pd.concat([selected, subs])
len(subsample)
```

*shuffling our data set*

Luckily, there are no missing values in our predictors and I, therefore, did not have to worry about missing value imputation.

The challenging part was: Creating a training data set that will allow our algorithms to pick up the specific characteristics that make a client likely to subscribe a term deposit as a result of this campaign. Using the original data set would not prove to be a good idea for this simple reason: Since over **89%** of our responders did not subscribe a term deposit, an algorithm that always predicts that the responder is non-subscriber would achieve an accuracy higher than **89%**. Nevertheless, that is the opposite of what we want. We do not want a **89%** accuracy that is achieved by never labeling a responder as a subscriber, we want to detect those who are likely to subscribe and label them as such.

There are two key points to focus on to help solve this. First, I used **random under-sampling** to create a training dataset with a balanced **y** distribution that will force the algorithms to detect subscribers as such to achieve high performance. Speaking of performance, I am going to make use of the **Receiver Operating Characteristics-Area Under the Curve** or **ROC-AUC** performance measure. Essentially, the **ROC-AUC** outputs a value between zero and one, whereby one is a perfect score and zero the worst. If an algorithm has a **ROC-AUC** score of above 0.5, it is achieving a higher performance than random guessing. 

To create a balanced training data set, I performed a split of the data into two samples (train 90% of the records and test 10%), I took all of the subscribing records in the training set ( there are **4203** subscription records in the train data). Then, I randomly selected the same number of non-subscribing records and concatenated the two.
